```
import numpy as np
import pandas as pd
from keras.utils import to_categorical
```

```
dataset = pd.read_csv('/content/datasets_13720_18513_insurance.csv')
dataset
```

```
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
dataset['region']=encoder.fit_transform(dataset['region'])
dataset['sex']=encoder.fit_transform(dataset['sex'])
dataset['smoker']=encoder.fit_transform(dataset['smoker'])

x = dataset.iloc[:,0:6].values
y = dataset.iloc[:,6].values

dataset
```

```
from sklearn.model_selection import train_test_split

X_train,X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2)

print(X_test)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Model initialization
regression_model = LinearRegression()
# Fit the data(train the model)
regression_model.fit(X_train, Y_train)
# Predict
y_predicted = regression_model.predict(X_test)
print(mean_absolute_error(Y_test,y_predicted))
```

```
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, Adam

model = Sequential()

model.add(Dense(50,input_shape=(6,),activation='tanh'))
model.add(Dense(30, activation='elu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))

model.compile(Adam(lr=0.005), loss='mean_absolute_error', metrics='accuracy')

model.summary()
```

```
from sklearn.metrics import mean_absolute_error,mean_squared_error, explained_variance_score
model.fit(X_train,Y_train,epochs = 500)
y_pred = model.predict(X_test)
print(mean_absolute_error(Y_test,y_pred))
print(explained_variance_score(Y_test,y_pred))
```
